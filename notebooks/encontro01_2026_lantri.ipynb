## Fundamentos de Programação

```python
import httpx
from bs4 import BeautifulSoup
import time # Adicionado para evitar bloqueios por excesso de velocidade

paginas = [
    "https://www.gov.br/mre/pt-br/canais_atendimento/imprensa/notas-a-imprensa/notas-a-imprensa?b_start:int=0",
    "https://www.gov.br/mre/pt-br/canais_atendimento/imprensa/notas-a-imprensa/notas-a-imprensa?b_start:int=30"
]

def acessa_pagina(link):
    print(f"Acessando: {link}")

    # Cabeçalhos para fingir ser um navegador real e assim não ser barrado dentro do site
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    }

    try:
        # Usamos follow_redirects=True para evitar erros de redirecionamento
        res = httpx.get(link, headers=headers, follow_redirects=True, timeout=20.0)
        res.raise_for_status()
        # Garante que a requisição deu certo (status 200)
        return BeautifulSoup(res.text, "html.parser")
    except Exception as e:
        print(f"Erro ao acessar {link}: {e}")
        return None

# próxima etapa para acessar a página inteira

for url in paginas: #acesso das páginas dentro dos links disponibilizados
    pagina_inteira = acessa_pagina(url)

    if not pagina_inteira: #se não carregar o link da página, pule para o próximo link
        continue

    content_core = pagina_inteira.find("div", attrs={"id": "content-core"})
    if not content_core:
        continue

    lista_noticias = content_core.find_all("article")

    for noticia in lista_noticias:
        # 1. Título
        titulo_tag = noticia.find("h2", attrs={"class": "tileHeadline"})
        titulo = titulo_tag.text.strip() if titulo_tag else "Sem título"
        print(titulo)

        # 2. Link
        link_noticia = noticia.a["href"] if noticia.a else "Sem link"
        print(link_noticia)

        # 3. Número da nota (subtítulo)
        sub = noticia.find("span", attrs={"class": "subtitle"})
        num_nota = sub.text.strip() if sub else "Sem número"
        print(num_nota)

        # 4. Data e Hora (Buscando nos ícones específicos)
        icones = noticia.find_all("span", attrs={"class": "summary-view-icon"})

        data_texto = "Data não encontrada"
        hora_texto = "Hora não encontrada"

        for item in icones:
            if item.find("i", class_="icon-day"):
                data_texto = item.get_text(strip=True)
            if item.find("i", class_="icon-hour"):
                hora_texto = item.get_text(strip=True)
                print(data_texto)
                print(hora_texto)

        # 5. Conteúdo da notícia

        conteudo_noticia = acessa_pagina(link_noticia)
        texto_conteudo = conteudo_noticia.find("div", attrs={"property": "rnews:articleBody"})
        if texto_conteudo:
          paragrafo = texto_conteudo.find_all("p")
          print(paragrafo)

        else:
          print("Não foi possível encontrar o conteúdo da notícia.")
          continue

        print("-" * 30)

    # Dica: adicione um pequeno intervalo para o servidor não te bloquear
    time.sleep(1)
```
