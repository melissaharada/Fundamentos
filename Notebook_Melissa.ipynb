```python
frequencia = contexto["PY"].value_counts(sort=False).tolist()
anos = contexto["PY"].astype(int).unique().tolist()
visualizar = {"quant artigos": frequencia, "anos": anos}
df = pd.DataFrame(visualizar)
df
```

```python
grafico_barra = df.plot(kind="barh", x="anos", y="quant artigos", color="pink")
```

```python
# ao colocar AB (somente  abstract), dá erro por não haver abstract em todos os textos da revista
palavra_chave = input("Escreva uma palavra-chave: ")
titulo = contexto[contexto["TI"].str.contains(palavra_chave)]
titulo
```

```python
# ao colocar AB (somente  abstract), dá erro por não haver abstract em todos os textos da revista
palavra_chave = input("Escreva uma palavra-chave: ")
busca = contexto[contexto["AB"].str.contains(palavra_chave,case=True,na=False)]
busca
```

```python
# & (ambos os termos na mesma busca/filtragem) e | (or, um dos termos presentes na filtragem)
coluna=contexto["AB"]
busca = contexto[coluna.str.contains("brasil",case=False,na=False)|coluna.str.contains("argentina",case=False,na=False)]
busca.shape
```
